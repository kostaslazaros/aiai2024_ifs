```python
import hdf5plugin
import scanpy as sc
import numpy as np
import pandas as pd
```

```python
# ADATA_PATH = "./data/4_concatenated_h5ad/lung_cancer_concat_data_v2.h5ad"

ADATA_PATH = "./data/4_concatenated_h5ad/lung_cancer_concat_data_v3.h5ad"
```

```python
merged_adata = sc.read_h5ad(ADATA_PATH)
merged_adata
```

```python
sc.pp.normalize_total(merged_adata, target_sum=1e4)
sc.pp.log1p(merged_adata)
sc.pp.scale(merged_adata, max_value=10)
```

```python
df = pd.DataFrame(merged_adata.X)
df
```

```python
df.columns = merged_adata.var.index
df
```

```python
df["tissue"] = merged_adata.obs["tissue"].values
df["cell_type"] = merged_adata.obs["cell_type"].values
```

```python
df["cell_type"].value_counts()
```

```python
merged_adata.obs["cell_type"].value_counts()
```

```python
# Reset the index and do not add the old index as a column
df = df.reset_index(drop=True)
df
```
